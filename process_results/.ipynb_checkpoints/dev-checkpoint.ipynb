```python
def process_matlab_output(a):
    try:
        sha = a.shape
    except:
        return a
    if len(sha)==2:
        m,n = sha
        if m==1 and n==1:
            return a[0][0]
        if m==1 or n==1:
            return a.reshape(-1)
        return a
    elif len(sha)==1:
        m = sha[0]
        if m==1:
            return a[0]
    else:
        return a
    
    
def abslog(a):
    return np.log10(np.abs(a))

def compute_vertices(rads, normalize=False):
    if normalize:
        rads = rads/np.mean(rads)
    n = len(rads)
    angles = np.arange(0,n)*np.pi/(n-1)
    c,s = np.cos(angles), np.sin(angles)
    return np.array([rads*c, rads*s])

def plot_vertices(verts):
    xs = list(verts[0])
    ys = list(verts[1])
    xs = xs+[xs[0]]
    ys = ys+[ys[0]]
    plt.plot(xs, ys)
    plt.axis('equal')
    return

def plot_semicircle():
    xs = np.linspace(-1,1,1000)
    ys = np.sqrt(1-xs**2)
    plot_vertices(np.array([xs,ys]))
    return

def compute_ellipse_rads(rx,ry,cx,cy,n):
    angles = np.arange(0,n)*np.pi/(n-1)
    cosine, sine = np.cos(angles), np.sin(angles)
    a = (ry**2)*(cosine**2)+(rx**2)*(sine**2)
    b = -2*cx*(ry**2)*cosine-2*cy*(rx**2)*sine
    c = (cx**2)*(ry**2)+(cy**2)*(rx**2)-(rx**2)*(ry**2)
    return (-b+np.sqrt(b**2-4*a*c))/(2*a)
```

```python
res = []
for i in range(1,6):
    fd = f'/home/zj/cluster/so0724/gd_ellipse{i}'
    fns = [i for i in os.listdir(fd) if i.endswith('.mat')]
    for fn in fns:
        data = loadmat(f'{fd}/{fn}')
        data['ellipse_idx'] = i
        res.append(data)
df = pd.DataFrame(res)
cols = [
    'init_rads', 'rads', 'f_zero', 'grad', 'iter_idx', 'ellipse_idx','min_step_size_reached', 'step','zk'
]
for c in cols:
    df[c] = df[c].apply(process_matlab_output)
df = df[cols].sort_values(['ellipse_idx', 'iter_idx']).reset_index(drop=True)
df['vertices'] = df.rads.apply(compute_vertices)
df['init_vertices'] = df.init_rads.apply(compute_vertices)
df['difference_from_optimal'] = df.f_zero-optimal_value
df['rads_from_semicircle'] = df.rads.apply(lambda x: np.linalg.norm(x-1))
df['grad_norm'] = df.grad.apply(lambda x: np.linalg.norm(x))
df_all = df.copy()
```

```python
ellipse_idxs = list(df_all.ellipse_idx.unique())
xlims = (-2,2)
ylims = (-0.1, 3.9)
fig, axs = plt.subplots(nrows = 3, ncols = 3, figsize = (10,10))
for i, ellipse_idx in enumerate(ellipse_idxs):
    r,c = divmod(i, 3)
    ax = axs[r][c]
    row = df.query(f"ellipse_idx=={ellipse_idx}").sort_values('iter_idx').iloc[0]
    
    verts = row.init_vertices
    xs = list(verts[0])
    ys = list(verts[1])
    xs = xs+[xs[0]]
    ys = ys+[ys[0]]
    ax.plot(xs, ys, label = f'init')
    
    verts = row.vertices
    xs = list(verts[0])
    ys = list(verts[1])
    xs = xs+[xs[0]]
    ys = ys+[ys[0]]
    ax.plot(xs, ys, label = f'first iter')
    
    ax.set_xlim(*xlims)
    ax.set_ylim(*ylims)
    ax.legend()
```

# Gradient convergence

```python
fd = f'/home/zj/shape_optimization_results/so0629'
res= []
for i in range(1,9):
    fns = os.listdir(f'{fd}/cvg{i}')
    for fn in fns:
        res.append(loadmat(f'{fd}/cvg{i}/{fn}'))
df = pd.DataFrame(res)
cols = [
    'h_power', 'h', 'half_n', 'grad', 'zk', 'f_plus', 'f_minus', 'f_zero'
]
df = df[cols]
for c in cols:
    df[c] = df[c].apply(process_matlab_output)
# df = df.query("h_power>2")
df['second_derivative'] = (df.f_plus-2*df.f_zero+df.f_minus)/((df.h)**2)
```

```python
fig, axes = plt.subplots(ncols = 2, nrows = 2, figsize = (15,8))
for n_idx, half_n in enumerate([8,16,32,64]):
    n = half_n*2
    row_idx, col_idx = divmod(n_idx, 2)
    axe = axes[row_idx][col_idx]
    df_plot = df[df.half_n==half_n].sort_values('h_power').reset_index(drop = True)
    h_powers = df_plot.h_power.to_list()
    grad = np.array(df_plot.grad.to_list())
    for i in range(half_n):
        axe.plot(h_powers, abslog(grad[:,i]))
    axe.set_title(f'{n} vertices')
    if col_idx == 0:
        axe.set_ylabel('log 10 partial derivative')
    if row_idx == 1:
        axe.set_xlabel('log 10 finite diff step size')
```
