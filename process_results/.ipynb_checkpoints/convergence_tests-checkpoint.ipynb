```python
import magic
import sys, os
import numpy as np
from matplotlib import pyplot as plt
from scipy.io import loadmat
import pandas as pd
from tqdm.notebook import tqdm
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
%matplotlib inline
import plotly.io as pio
pio.renderers.default = 'notebook'
```

```python
optimal = 0.3655840228073865
optimal_value = -optimal
optimal_zk = 3.831705970207512
optimal_normal_derivative = 5.367493890087601
```

```python
def process_matlab_output(a):
    m,n = a.shape
    if m==1 and n==1:
        return a[0][0]
    if m==1 or n==1:
        return a.reshape(-1)
    return a

def abslog(a):
    return np.log10(np.abs(a))
```

# shape convergence

```python
def compute_vertices(half_rads):
    n = len(half_rads)*2
    rads = np.array(list(half_rads) + list(half_rads)[::-1])
    angles = np.arange(0,n)*np.pi/(n-1)
    c = np.cos(angles)
    s = np.sin(angles)
    return np.array(
        [c*rads, s*rads]
    )
```

```python
fd = f'/home/zj/shape_optimization_results/so0628'
res= []
for i in range(1,3):
    fns = os.listdir(f'{fd}/gd{i}')
    for fn in fns:
        res.append(loadmat(f'{fd}/gd{i}/{fn}'))
df = pd.DataFrame(res)
cols = [
    ''
]
cols = [
    'n', 'h', 'rads', 'zk', 'min_step_size_reached', 'max_iter_reached', 'not_converged', 'iter_idx', 'f_zero',
    'grad', 'f_plus', 'f_minus', 'improvement', 'step'
]
df = df[cols]
for c in cols:
    df[c] = df[c].apply(process_matlab_output)
# df = df.query("h_power>2")
df['second_derivative'] = (df.f_plus-2*df.f_zero+df.f_minus)/((df.h)**2)
df['vertices'] = df.rads.apply(compute_vertices)
df['difference_from_optimal'] = df.f_zero - optimal_value
df['grad_norm'] = df.grad.apply(np.linalg.norm)
df['dx1'] = df.grad.apply(lambda x: x[0])
df['dx2'] = df.grad.apply(lambda x: x[1])
df['dx_mid'] = df.grad.apply(lambda x: x[-1])
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 5, figsize = (15,3))
axe = axes[0]
for n in [8, 64]:
    df_plot = df[df.n==n].sort_values('iter_idx').reset_index(drop = True)
    axe.plot(abslog(df_plot.difference_from_optimal), label = f'{n} vertices')
axe.set_xlabel('number of iterations')
axe.set_ylabel('log10 difference from optimal')
axe.legend()

axe = axes[1]
for n in [8, 64]:
    df_plot = df[df.n==n].sort_values('iter_idx').reset_index(drop = True)
    axe.plot(abslog(df_plot.grad_norm), label = f'{n} vertices')
axe.set_xlabel('number of iterations')
axe.set_ylabel('log10 gradient norm')
axe.legend()

axe = axes[2]
for n in [8, 64]:
    df_plot = df[df.n==n].sort_values('iter_idx').reset_index(drop = True)
    axe.plot(abslog(df_plot.dx1), label = f'{n} vertices')
axe.set_xlabel('number of iterations')
axe.set_ylabel('log10 dx1')
axe.legend()

axe = axes[3]
for n in [8, 64]:
    df_plot = df[df.n==n].sort_values('iter_idx').reset_index(drop = True)
    axe.plot(abslog(df_plot.dx2), label = f'{n} vertices')
axe.set_xlabel('number of iterations')
axe.set_ylabel('log10 dx2')
axe.legend()
axe = axes[4]
for n in [8, 64]:
    df_plot = df[df.n==n].sort_values('iter_idx').reset_index(drop = True)
    axe.plot(abslog(df_plot.dx_mid), label = f'{n} vertices')
axe.set_xlabel('number of iterations')
axe.set_ylabel('log10 dx mid')
axe.legend()
```

# Objective convergence

```python
fd = f'/home/zj/shape_optimization_results/so0630'
fns = os.listdir(fd)
res= []
for fn in tqdm(fns):
    res.append(loadmat(f'{fd}/{fn}'))
df = pd.DataFrame(res)
cols = [
    'half_n', 'f_zero', 'ud_p', 'uq', 'zk'
]
df = df[cols]
for c in cols:
    df[c] = df[c].apply(process_matlab_output)
df = df.sort_values('half_n').drop_duplicates(subset = ['half_n']).reset_index(drop=True)
```

```python
df['distance_from_optimal'] = df.f_zero - optimal_value
df['number_of_vertices'] = ((df.half_n)*2.0).astype(int)
df['normal_derivative'] = df.ud_p/df.uq
df['distance_from_zk'] = df.zk - optimal_zk
df['distance_from_normal_derivative'] = df.normal_derivative - optimal_normal_derivative
```

```python
min_verts = 5
fig = make_subplots(rows=3, cols=1, subplot_titles=("normal derivative (polygon minus semicircle)", "zk (polygon minus semicircle)", "objective (polygon minus semicircle)"))
df_plot = df.query(f"half_n>={min_verts}")
fig.add_trace(
    go.Scatter(x=df_plot['number_of_vertices'], y=df_plot['distance_from_normal_derivative']),
    row=1, col=1)
fig.update_yaxes(tickformat = 'e', type='log', row=1, col=1)
fig.add_trace(
    go.Scatter(x=df_plot['number_of_vertices'], y=df_plot['distance_from_zk']),
    row=2, col=1)
fig.update_yaxes(tickformat = 'e', type='log', row=2, col=1)
fig.add_trace(
    go.Scatter(x=df_plot['number_of_vertices'], y=df_plot['distance_from_optimal']),
    row=3, col=1)
fig.update_yaxes(tickformat = 'e', row=3, col=1)
fig.update_layout(
    autosize=False,
    width=800,
    height=800)
fig.show()
```

# Gradient convergence

```python
fd = f'/home/zj/shape_optimization_results/so0629'
res= []
for i in range(1,9):
    fns = os.listdir(f'{fd}/cvg{i}')
    for fn in fns:
        res.append(loadmat(f'{fd}/cvg{i}/{fn}'))
df = pd.DataFrame(res)
cols = [
    'h_power', 'h', 'half_n', 'grad', 'zk', 'f_plus', 'f_minus', 'f_zero'
]
df = df[cols]
for c in cols:
    df[c] = df[c].apply(process_matlab_output)
# df = df.query("h_power>2")
df['second_derivative'] = (df.f_plus-2*df.f_zero+df.f_minus)/((df.h)**2)
```

```python
fig, axes = plt.subplots(ncols = 2, nrows = 2, figsize = (15,8))
for n_idx, half_n in enumerate([8,16,32,64]):
    n = half_n*2
    row_idx, col_idx = divmod(n_idx, 2)
    axe = axes[row_idx][col_idx]
    df_plot = df[df.half_n==half_n].sort_values('h_power').reset_index(drop = True)
    h_powers = df_plot.h_power.to_list()
    grad = np.array(df_plot.grad.to_list())
    for i in range(half_n):
        axe.plot(h_powers, abslog(grad[:,i]))
    axe.set_title(f'{n} vertices')
    if col_idx == 0:
        axe.set_ylabel('log 10 partial derivative')
    if row_idx == 1:
        axe.set_xlabel('log 10 finite diff step size')
```

```python
fig, axes = plt.subplots(ncols = 2, nrows = 2, figsize = (15,8))
for n_idx, half_n in enumerate([8,16,32,64]):
    n = half_n*2
    row_idx, col_idx = divmod(n_idx, 2)
    axe = axes[row_idx][col_idx]
    df_plot = df[df.half_n==half_n].sort_values('h_power').reset_index(drop = True)
    h_powers = df_plot.h_power.to_list()
    fdd = np.array(df_plot.second_derivative.to_list())
    axe.plot(h_powers, abslog(fdd))
    axe.set_title(f'{n} vertices')
    if col_idx == 0:
        axe.set_ylabel('log 10 second deriv along grad direction')
    if row_idx == 1:
        axe.set_xlabel('log 10 finite diff step size')
```
